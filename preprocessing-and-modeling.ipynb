# Car price predictor using Regression

```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

from sklearn.ensemble import ExtraTreesRegressor
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import r2_score

import pickle
```

```python
# data importing
df = pd.read_csv('data/data.csv')

# script to show all columns
pd.set_option('display.max_columns', None)

# suppress scientific format
pd.set_option('display.float_format', lambda x: '%.5f' % x)

df.head()
```

```python
df.shape
```

```python
df.info()
```

## Preprocessing

```python
# Searching for unique values from the object-typed features
print(df['fuel'].unique())
print(df['seller_type'].unique())
print(df['transmission'].unique())
print(df['owner'].unique())
```

```python
# Creating feature of years old
df['years_old']=2020-df['year']
```

```python
# Preprocessing of object-typed features (except "name" feature)
df=pd.get_dummies(data=df,columns=['fuel','seller_type','transmission','owner'])
```

```python
# Drop not used features
df.drop(['name','year'],axis=1,inplace=True)
df.head()
```

```python
# Searching for correlations between features
df.corr()
```

```python
corr_val=df.corr()
corr_features=corr_val.index
plt.figure(figsize=(10,10))
sns.heatmap(df[corr_features].corr(),annot=True,cmap="RdYlGn")
```

## Building model parameters

```python
# Define training and target features
X = df.iloc[:,1:]
y = df.iloc[:,0]

# Show training df
X.head()
```

```python
list(X)
```

```python
# Rename X features with spaces in their names
X.rename(columns={'seller_type_Trustmark Dealer': 'seller_type_TrustmarkDealer', 
                  'owner_First Owner': 'owner_First_Owner',
                  'owner_Fourth & Above Owner': 'owner_Fourth_Above_Owner',
                  'owner_Second Owner': 'owner_Second_Owner',
                  'owner_Test Drive Car': 'owner_Test_Drive_Car',
                  'owner_Third Owner': 'owner_Third_Owner',}, inplace=True)
```

```python
list(X)
```

```python
# Show selling prices
y.head()
```

```python
# Determine feature importance
model = ExtraTreesRegressor()
model.fit(X,y)

# plot feature importances
feat_imp = pd.Series(model.feature_importances_, index=X.columns)
feat_imp.nlargest(19).plot(kind='barh')
plt.show()
```

#### The graph shows us that the "years_old" and "km_driven" features are the ones with the highest incidence in the price, followed by the transmission features and if it is powered by diesel.

## Building the model

```python
# Splitting the data into test and train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
X_train.shape, X_test.shape
```

```python
# Hyperparameters
n_estimators = [int(x) for x in np.linspace(start=100, stop=1200, num=12)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(5, 30, num=6)]
min_samples_split = [2, 5, 10, 15, 100]
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Create a random grid of hyperparameters for RandomizedSearchCV
random_grid = {'n_estimators': n_estimators,
              'max_features': max_features,
              'max_depth': max_depth,
              'min_samples_split': min_samples_split,
              'min_samples_leaf': min_samples_leaf}
print(random_grid)
```

```python
# Building the model
rfr = RandomForestRegressor()
```

```python
rfr_random = RandomizedSearchCV(estimator = rfr, param_distributions=random_grid, 
                                scoring='neg_mean_squared_error', 
                                n_iter = 10, cv = 5, verbose = 2, 
                                random_state=42, n_jobs = 1)
```

```python
# Train the model
rfr_random.fit(X_train, y_train)
```

```python
# Predict
y_pred = rfr_random.predict(X_test)
y_pred
```

## Evaluate the model

```python
sns.distplot(y_test-y_pred)
```

```python
plt.scatter(y_test,y_pred)
```

```python
print("R2 score: ", r2_score(y_test, y_pred))
```

## Save the model

```python
#open a file to store the data
file = open('rfr_model.pkl', 'wb')

# Dump information to that file
pickle.dump(rfr_random, file)

print("Done!")
```
